```python
import torchvision
import torch
import torchvision.transforms as T
from torchvision.models.detection.faster_rcnn import FastRCNNPredictor
import matplotlib.pyplot as plt
import os
from PIL import Image
import cv2
from xml.dom.minidom import parse
import numpy as np 
#import utils
from engine import train_one_epoch, evaluate
```

```python
from glob import glob
import os
def get_image_names(image_f):
    files = os.listdir(image_f)
    return files

names = get_image_names('NNData')
```

```python
def process_image(name):
    bbox_xml_path = os.path.join('Annotations', name.split('.')[0] + '.xml')
    
    dom = parse(bbox_xml_path)                        
    data = dom.documentElement
    objects = data.getElementsByTagName('object')        

    boxes = []
    labels = []
    for object_ in objects:
        name = object_.getElementsByTagName('name')[0].childNodes[0].nodeValue
        if name == "Galaxy":
            labeller = 1        
        elif name == "Cavity":
            labeller = 2
        else:
            labeller = 0 
        labels.append(labeller)  

        bndbox = object_.getElementsByTagName('bndbox')[0]
        xmin = np.float(bndbox.getElementsByTagName('xmin')[0].childNodes[0].nodeValue)        
        ymin = np.float(bndbox.getElementsByTagName('ymin')[0].childNodes[0].nodeValue)
        xmax = np.float(bndbox.getElementsByTagName('xmax')[0].childNodes[0].nodeValue)
        ymax = np.float(bndbox.getElementsByTagName('ymax')[0].childNodes[0].nodeValue)
        boxes.append([xmin, ymin, xmax, ymax])        

    return boxes, labels
```

```python
def visualize_image(boxes, labels, name, plot = False):
    img = cv2.imread(os.path.join('NNData', name))
    for b,l in zip(boxes, labels):
        if (l==1):
            colour = (255,0,0)
        elif(l==2):
            colour = (0,0,255)
        cv2.rectangle(img, (int(b[0]), int(b[1])), (int(b[2]),int(b[3])), colour, 2)
    if(plot == True):
        plt.figure(figsize = (20,10))
        plt.imshow(img)
        plt.title(name)
        plt.show()
        return
    return img

```

```python
b, l = process_image(names[35])

visualize_image(b, l, names[35], True)

# red = galaxy
# blue = cavity
```

```python
def visualize_many():
    fig, axs = plt.subplots(3,3, figsize = (20,20))
    imgs = []
    titles = []
    for i in range(10,19):
        b, l = process_image(names[i])
        img = visualize_image(b, l, names[i])    
        imgs.append(img)
        titles.append(names[i])
        
    axs[0, 0].imshow(imgs[0])
    axs[0, 1].imshow(imgs[1])
    axs[0, 2].imshow(imgs[2])
    axs[1, 0].imshow(imgs[3])
    axs[1, 1].imshow(imgs[4])
    axs[1, 2].imshow(imgs[5])
    axs[2, 0].imshow(imgs[6])
    axs[2, 1].imshow(imgs[7])
    axs[2, 2].imshow(imgs[8])
    #axs[0, 0].imshow(imgs[9])
    
    
    
    fig.tight_layout()
```

```python
visualize_many()
```

## DILATION/EROSION EXPERIMENTATION

```python
idx = 20
b, l = process_image(names[idx])
visualize_image(b, l, names[idx], True)
```

```python
def generate_new_image(name, sample = False):
    img = cv2.imread(os.path.join('NNData', name))
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    kernel = np.ones((5, 5), 'uint8')

    dilate_img = cv2.dilate(gray, kernel, iterations=3)
    erode_img = cv2.erode(gray, kernel, iterations=3)
    
    if(sample == True):
        return img, gray, dilate_img, erode_img
    
    output_image = cv2.merge ( (gray, dilate_img, erode_img) )
    return output_image
```

```python
i,g,d,e = generate_new_image(names[idx], True)
```

```python
fig, axs = plt.subplots(2,2, figsize = (30,20))
axs[0,0].imshow(i)
axs[0,0].set_title('original')

axs[0,1].imshow(g, cmap = 'gray')
axs[0,1].set_title('gray')

axs[1,0].imshow(d, cmap = 'gray')
axs[1,0].set_title('dilated')

axs[1,1].imshow(e, cmap = 'gray')
axs[1,1].set_title('eroded')



```

```python
new = generate_new_image(names[idx])
plt.figure(figsize = (10,10))
plt.imshow(new, cmap = 'gray')
```

```python
## generate the new images and save them

for n in names:
    new = generate_new_image(n)
    new_name = n.split('.')[0] + '_aug_.png'
    folder = 'augmented_img/'
    
    cv2.imwrite(folder+new_name, new)
    print(n)
```
